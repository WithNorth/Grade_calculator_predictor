```python
class FifthWindow:
    def __init__(self, master, selected_weekly_data, file_path, num_weeks):
        self.master = master
        self.file_path = file_path
        self.selected_weekly_data = selected_weekly_data

        self.num_weeks = num_weeks
        self.data_frame = self.read_csv_columns()

        self.filename = os.path.basename(file_path)
        self.filename_label = tk.Label(master, text= self.filename + " Result\n")
        self.filename_label.pack(side=tk.TOP)
        self.results_text = scrolledtext.ScrolledText(master, height=10, width=50)
        self.results_text.pack(pady=10)

        self.calculate_and_display_results()

        self.new_file_button = tk.Button(master, text="New File", command=self.new_file)
        self.new_file_button.pack(side=tk.LEFT, anchor=tk.SW, padx=10, pady=10)

        self.exit_button = tk.Button(master, text="Exit", command=self.exit)
        self.exit_button.pack(side=tk.RIGHT, anchor=tk.SE, padx=10, pady=10)

    
    def exit(self):
        self.master.destroy()


    def new_file(self):
        self.selected_weekly_data = None
        self.file_path = None
        self.num_weeks = None
        self.data_frame = None
        print("The past data has been cleared!!!")
        # clear the previous data after saving them in a csv file
        self.master.destroy()
        create_second_window()

    def calculate_and_display_results(self):
        weeks = []
        mean_squared_err = []
        r2score = []

        for week, columns in self.selected_weekly_data.items():
            # global selected_columns  do not use global otherwise cleaning data is needed to clarify global as well
            selected_columns=[]
            for column in columns:
                if isinstance(column,str) and column:
                    selected_columns.append(column)
            selected_columns = list(set(selected_columns))
            print("Processing Week:", week)

            print(f"All selected columns till now: {selected_columns}")
            mse, r2 = self.calculate_metrics(selected_columns, week)

            weeks.append(f'Week{week}')
            mean_squared_err.append(mse)
            r2score.append(r2)


        self.display_results(weeks, mean_squared_err, r2score)
        self.save_results(weeks, mean_squared_err, r2score)

    def save_results(self, weeks,mean_squared_err,r2score):
        result_df = pd.DataFrame({
            'Weeks' : weeks,
            'mean squared error' : mean_squared_err,
            'R2-Score': r2score
        })
        # output_path = 'd:\study\Fall 2023\cs education\ result1.csv'
        # Finished try, directly open that file to show
        # The results are the same as we dealt with codes
        # output_path = 'd:\study\Fall 2023\cs education\ result2.csv'
        # This path has already saved the data with the whole selection
        # output_path = 'd:\study\Fall 2023\cs education\ result3.csv'
        # This path is used to check whether the last file selection data has been cleared or not
        # Fixed 
        output_path = 'D:\study\Spring 2024\cs education\combined_part\predictor\ R2_result.csv'
        # To display
        result_df.to_csv(output_path)

    def calculate_metrics(self, columns, num_weeks):
            column_order = list(columns)
            with open(f'D:\study\Spring 2024\cs education\combined_part\calculator\column_order\ column_order_{num_weeks}.json','w') as f:
                json.dump(column_order,f)
            print("Selected Columns:", columns) 
            print()
            
            if not columns:
                print(f"No columns selected for week {num_weeks}")
                return float('NaN'), float('NaN')

            independent_variable = self.data_frame[columns]
            
            # Final grade as the dependent variable
            dependent_variable = self.data_frame.iloc[:, -1]
            
            X_train, X_test, y_train, y_test = model_selection.train_test_split(independent_variable, dependent_variable, test_size=0.2, random_state=135153)
            linear_reg = LinearRegression()
            linear_reg.fit(X_train, y_train)

            model_file = f'D:\study\Spring 2024\cs education\combined_part\predictor\model_stored\ model_week_{num_weeks}.pkl'
            with open(model_file, 'wb') as file:
                pickle.dump(linear_reg, file)
            print(f'Model for week {num_weeks} saved to {model_file}\n')

            pred = linear_reg.predict(X_test)

            mse = round(mean_squared_error(y_test, pred),2)
            r2 = round(r2_score(y_test, pred),3)
            print(f"Week {num_weeks}, MSE: {mse}, R2: {r2}")

            return mse, r2
    


    def display_results(self, weeks, mse, r2):
        self.results_text.config(state=tk.NORMAL)
        for week, mse_val, r2_val in zip(weeks, mse, r2):
            self.results_text.insert(tk.END, f"{week}: MSE={mse_val:.2f}, R2={r2_val:.2f}\n")
        self.results_text.config(state=tk.DISABLED)



    def read_csv_columns(self):
        self.data_frame = pd.read_csv(self.file_path)
        print("DataFrame Columns:", self.data_frame.columns.tolist())

        return self.data_frame
        # return self.data_frame.columns.tolist()


def create_fifth_window(file_path,selected_weekly_data, num_weeks):
    fifth_window_root = tk.Tk()
    FifthWindow(fifth_window_root, selected_weekly_data, file_path,num_weeks)
    fifth_window_root.mainloop()
```

```python
class ThirdWindow:
    def __init__(self, master, file_path, num_weeks = 16):
        # "num_weeks" should be set as the week nums you wanted, but it should show the exact number
        self.master = master
        self.filename = os.path.basename(file_path)
        self.file_path = file_path
        self.csv_columns = self.read_csv_columns()
        # self.all_columns = pd.DataFrame({'columns': self.csv_columns})

        self.num_weeks = num_weeks
        self.current_week = 1
        self.current_page = 1
        self.max_pages = (len(self.csv_columns) - 1) // 8 + 1  # calculate the total pages
        # self.checkbox_states = {}  
        # self.checked_columns = set()
        self.weekly_checkbox_states = {week: {} for week in range(1, num_weeks + 1)}
        self.current_week_checkbox_states = self.weekly_checkbox_states[self.current_week]

        self.filename_label = tk.Label(master, text="Filename: " + self.filename)
        self.filename_label.pack()

        self.week_label = tk.Label(master, text="Week " + str(self.current_week))
        self.week_label.pack()

        self.checkboxes_frame = tk.Frame(master)
        self.checkboxes_frame.pack()

        self.page_label = tk.Label(master, text=f"Page {self.current_page}/{self.max_pages}")
        self.page_label.pack()

        self.prev_button = tk.Button(master, text="Previous", command=self.prev_page)
        self.prev_button.pack(side=tk.LEFT)
        self.prev_button.config(state=tk.DISABLED)

        self.next_button = tk.Button(master, text="Next", command=self.next_page)
        self.next_button.pack(side=tk.LEFT)
        if self.max_pages == 1:
            self.next_button.config(state=tk.DISABLED)

        self.next_week_button = tk.Button(master, text="Next Week", command=self.next_week)
        self.next_week_button.pack(side=tk.LEFT)

        self.finish_button = tk.Button(master, text="Finish Choosing", command=self.finish_choosing)
        self.finish_button.pack(side=tk.RIGHT)

        self.create_checkboxes()


    def create_checkboxes(self):
        for widget in self.checkboxes_frame.winfo_children():
            widget.destroy()

        for column_name in self.csv_columns:
            if column_name not in self.current_week_checkbox_states:
                self.current_week_checkbox_states[column_name] = tk.BooleanVar(value=False)

        available_columns = [col for col in self.csv_columns if not self.column_was_checked_in_previous_weeks(col)]

        self.max_pages = max(1, (len(available_columns) - 1) // 8 + 1)
        self.current_page = min(self.current_page, self.max_pages)

        page_start = (self.current_page - 1) * 8
        page_end = min(page_start + 8, len(available_columns))
        for i in range(page_start, page_end):
            column_name = available_columns[i]
            var = self.current_week_checkbox_states[column_name]

            cb = tk.Checkbutton(self.checkboxes_frame, text=column_name, variable=var)
            cb.pack(anchor=tk.W)

        self.update_page_controls()

    def column_was_checked_in_previous_weeks(self, column_name):
        return any(
            self.weekly_checkbox_states[week].get(column_name, False)
            for week in range(1, self.current_week)
        )
    
    def update_page_controls(self):
        self.page_label.config(text=f"Page {self.current_page}/{self.max_pages}")

        self.prev_button.config(state=tk.NORMAL if self.current_page > 1 else tk.DISABLED)
        self.next_button.config(state=tk.NORMAL if self.current_page < self.max_pages else tk.DISABLED)
        

    def prev_page(self):
        if self.current_page > 1:
            self.current_page -= 1
            self.create_checkboxes()

    def next_page(self):
        if self.current_page < self.max_pages:
            self.current_page += 1
            self.create_checkboxes()

 
    def next_week(self):
        self.weekly_checkbox_states[self.current_week] = {
            column_name: var.get() for column_name, var in self.current_week_checkbox_states.items()
        }

        if self.current_week < self.num_weeks:
            self.current_week += 1
            self.current_page = 1

            if self.current_week not in self.weekly_checkbox_states:
                self.weekly_checkbox_states[self.current_week] = {}
                for column_name in self.csv_columns:
                    self.weekly_checkbox_states[self.current_week][column_name] = tk.BooleanVar(value=False)
            else:
                for column_name, var in self.weekly_checkbox_states[self.current_week].items():
                    if column_name in self.current_week_checkbox_states:
                        self.current_week_checkbox_states[column_name].set(var)

            self.week_label.config(text="Week " + str(self.current_week))
            self.create_checkboxes()


    def finish_choosing(self):
        try:
            self.weekly_checkbox_states[self.current_week] = {
                column_name: var.get() for column_name, var in self.current_week_checkbox_states.items()
            } # needed to save the status of all the weeks selected
            
            selected_weekly_data = {
                week: [col for col, var in states.items() if var]
                for week, states in self.weekly_checkbox_states.items()
            }

            self.master.destroy()
            create_fourth_window(self.file_path,selected_weekly_data, self.num_weeks)

        except Exception as e:
            print(f"Error in finish_choosing: {e}")





    def read_csv_columns(self):
        with open(self.file_path, newline='') as csvfile:
            reader = csv.reader(csvfile)
            headers = next(reader)
            return headers   


def create_third_window(file_path):
    third_window_root = tk.Tk()
    third_window_root.title("Column Selection")
    ThirdWindow(third_window_root, file_path)
    third_window_root.mainloop()

```
